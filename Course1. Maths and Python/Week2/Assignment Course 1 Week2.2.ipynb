```python
import numpy as np
from matplotlib import pylab as plt
x = np.arange(1, 15, 0.1)
y = np.exp(x/10.0)*np.sin(x/5.0) + np.exp(-x/2.0)*5
plt.plot(x, y)
```

```python
def f(x):
    return np.exp(x/10.0)*np.sin(x/5.0) + np.exp(-x/2.0)*5
```

```python
y1=f(1)
y2=f(15)
A=np.array([[1., 1.], [1., 15.]], dtype=np.float32)
b=np.array([[y1],[y2]], dtype=np.float32)
print(b)
```

```python
S= np.linalg.solve(A, b)
print(S)
```

```python
x = np.arange(1, 15, 0.1)
y=S[0]+S[1]*x
plt.plot(x, y)
x = np.arange(1, 15, 0.1)
y = np.exp(x/10.0)*np.sin(x/5.0) + np.exp(-x/2.0)*5
plt.plot(x, y)
plt.show()
```

```python
y1=f(1)
y2=f(8)
y3=f(15)

A=np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]], dtype=np.float32)
b=np.array([[y1],[y2], [y3]], dtype=np.float32)
print(A)
print(b)

S= np.linalg.solve(A, b)
print(S)

x = np.arange(1, 15, 0.1)
y=S[0]+S[1]*x+S[2]*x**2
plt.plot(x, y)
x = np.arange(1, 15, 0.1)
y = np.exp(x/10.0)*np.sin(x/5.0) + np.exp(-x/2.0)*5
plt.plot(x, y)
plt.show()
```
